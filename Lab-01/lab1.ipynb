```python
# Create a csv file where there are 100 rows and two columns. 
# First column contains review, second column contains the label(Positive, negative, neutral).

print("Load the given CSV file containing text and label columns into a Pandas DataFrame.")

import pandas as pd

# https://www.kaggle.com/code/akanksha10/sentiment-analysis-dataset/input
df = pd.read_csv('test.csv', encoding='latin1')
 
df = df.drop(columns=[col for col in df.columns if col not in ['text', 'sentiment']])

df = df.dropna()
df = df.drop(df.index[100:])
df = df.rename(columns={'sentiment':'label'})
```

```python
print("Display the first three records of the dataset and print the total number of rows and columns.")
print(df.head(3))
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
```

```python
print("3. Print the column names and data types of each column.")
print(df.dtypes)
```

```python
print("4. Check whether the dataset contains any missing values and display the count of missing values per column.")
print(df.isnull().sum())
```

```python
print("5. Print the number of unique labels present in the dataset.")
print(df['label'].nunique())
```

```python
print("6. Display the count of records for each label.")
print(df['label'].value_counts())
```

```python
print("7. Filter and display all rows where the label is \"Negative\"")
negative_rows = df[df['label'] == 'negative']
print(negative_rows)
```

```python
print("8. Create a new column that stores the number of characters present in each text.")
df['number_of_characters'] = df['text'].apply(len)
print(df.head())
```

```python
print("9. Create another column that stores the number of words in each text.")
df['number_of_words'] = df['text'].apply(lambda x: len(x.split()))
print(df.head())
```

```python
print("10. Convert all text entries to lowercase and store the result in a new column without modifying the original text column.")
df['text_lower'] = df['text'].str.lower()
print(df.head())
```

```python
print("11. Standardize the label column by converting all labels to lowercase.")
df['label'] = df['label'].str.lower()
print(df.head())
```

```python
print("12. Create a new column in which the labels are encoded as follows: positive → 1, neutral → 0, negative → -1.")
df['sentiment_label'] = df['label'].map({'positive': 1, 'negative': -1, 'neutral': 0})
print(df.head())
```

```python
print("13. Sort the DataFrame in descending order based on the word count column.")
df = df.sort_values(by='number_of_words', ascending=False)
print(df)
```

```python
print("14. Display the top three longest text entries based on word count.")
print(df.head(3))
```

```python
print("15. Filter and display all texts that contain more than six words and belong to the \"positive\" label.")
print(df[df['number_of_words'] > 6][df['label'] == 'positive'])
```

```python
print("16. Group the data by label and compute the average word count for each group.")
grouped_df = df.groupby('label').agg({'number_of_words': 'mean'}).reset_index()
print(grouped_df)
```

```python
print("17. Identify the label that has the highest average word count.")
print(grouped_df['label'][grouped_df['number_of_words'].idxmax()])
```

```python
print("18. Count the number of text entries with fewer than five words for each label.")
count_df = df[df['number_of_words'] > 5].groupby('label').size().reset_index(name='count')
print(count_df)
```

```python
print("19. Change the DataFrame index to start from 1 instead of 0.")
df.index = df.index + 1
print(df)
```

```python
print("20. Rename the columns text to review_text and label to sentiment.")
df = df.rename(columns={'text': 'review', 'label': 'sentiment'})
print(df)
```

```python
print("21. Update the sentiment of a specific record (given index) from \"neutral\" to \"positive\".")
df['sentiment'][df['sentiment'] == 'neutral'][0] = 'positive'
print(df.head())
```

```python
print("22. Create a new DataFrame that contains only the rows with positive sentiment.")
positve_df = df[df['sentiment'] == 'positive']
print(positve_df)
```

```python
print("23. Save the cleaned and updated DataFrame to a new CSV file.")
positve_df.to_csv('cleaned_test.csv', index=False)
```

```python
print("24. Reload the saved file and verify that the changes are correctly reflected.")
df = pd.read_csv('cleaned_test.csv')
print(df.head())
```
